```python
import pandas as pd
import numpy as np
s = pd.Series(np.random.rand(10).cumsum(), index = np.arange(0,100, 10))
s.plot()
```

```python
import pandas as pd
import numpy as np
a = np.array([[1,2,3],[4,5,6]]).cumsum(0)
b = pd.DataFrame(np.random.rand(20).reshape(4,5).cumsum(0),
                 columns = pd.Index(['A', 'B', 'C', 'D', 'E'], name = 'random')
                 )
b.plot()
# vertical bar chart
b.plot.bar(color = 'k', alpha = 0.7)
# Horizontal bar chart
b.plot.barh()
b.plot.barh(stacked = True, alpha = 0.5)
```

```python
# Sea borne as sns Dist plot
import numpy as np
import seaborn as sns

# selecting style as white
sns.set(style = "white")

# generate a random univariate dataset
rs = np.random.RandomState(10)
d = rs.normal(size = 100)
c = np.random.rand(100)

# plot a simple histogram and kde with binsize selected automatically
sns.distplot(d,kde = True, color = 'm')
```

```python
# seaborn histogram distplot kde numpy concatenate
from operator import concat
import numpy as np
import pandas as pd
import seaborn as sns
c1 = np.random.normal(0,1,size = 100)
c2 = np.random.normal(10,1, size = 100)
values = pd.Series(np.concatenate([c1,c2]))
sns.distplot(values, kde = True, bins= 10, color = 'k')
```
