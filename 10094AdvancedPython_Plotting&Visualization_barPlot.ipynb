```python
import pandas as pd
import numpy as np
s = pd.Series(np.random.rand(10).cumsum(), index = np.arange(0,100, 10))
s.plot()
```

```python
import pandas as pd
import numpy as np
a = np.array([[1,2,3],[4,5,6]]).cumsum(0)
b = pd.DataFrame(np.random.rand(20).reshape(4,5).cumsum(0),
                 columns = pd.Index(['A', 'B', 'C', 'D', 'E'], name = 'random')
                 )
b.plot()
# vertical bar chart
b.plot.bar(color = 'k', alpha = 0.7)
# Horizontal bar chart
b.plot.barh()
b.plot.barh(stacked = True, alpha = 0.5)
```
