```python
# groupby and transform
import numpy as np
import pandas as pd

# Create a dataframe
df = pd.DataFrame({ 'Dept': np.random.choice(["IT", 'HR', 'SALES'], size = 10),
                   'Gender': np.random.choice(["F", "M"], size = 10),
                   'Age': np.random.randint(20, 60, size =10),
                   'Salary': np.random.randint(20000, 290000, size = 10),
                  })
df.index.name = "emp_id"
# calculate the mean difference value by transforming each salary value
df['mean_sal_diff'] = df['Salary'].groupby(df['Dept']).transform(lambda x: x - x.mean())
df.head()
```
