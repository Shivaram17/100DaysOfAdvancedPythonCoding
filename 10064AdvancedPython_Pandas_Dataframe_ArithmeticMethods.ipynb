```python
import pandas as pd
data = {'State': ['Chennai', 'Texas', 'Delhi', 'Hyderabad', 'Ontario'],
        'Year': [2000, 2001, 2002, 2003, 2004],
        'pop': [1,2,3,4,5]
        }

frame = pd.DataFrame(data)
frame.loc[1]

# add new column
frame2 = pd.DataFrame(data, columns = ['State', 'Year', 'pop', 'Accounting'],
                      index = ['one', 'two', 'three', 'four', 'five'])
frame2.Accounting = 'Citi Bank'
frame2
```

```python
import pandas as pd
import numpy as np
obj = pd.DataFrame(np.arange(9).reshape(3,3),
                      index = ['a', 'c', 'd'],
                      columns = ['Delhi', 'Chennai', 'Bangalore'])
obj.loc['a', ['Chennai', 'Delhi']]
# filtering using iloc for integers
obj.iloc[1,[1,2]]
obj.loc[:'d', 'Chennai':]
```

```python
# arithmetic methods using pandas
import pandas as pd
df1 = pd.DataFrame(np.arange(16).reshape(4,4),
                   index=['a','b', 'c', 'd'],
                   columns = list('abcd'))

df2 = pd.DataFrame(np.arange(25).reshape(5,5),
                   index = ['a','b','c','d','e'],
                   columns = list('abcde'))
df2.loc['a', 'b'] = np.nan
df1 + df2
df1.add(df2, fill_value = 0)
```
