```python
# matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x = pd.DataFrame(np.arange(20).reshape(4,5))
data = pd.Series(np.arange(15))
y = np.arange(10)
plt.plot(data)
plt.show()
plt.plot(y)
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
plt.show()
plt.plot(np.random.randn(50).cumsum(), 'k--')
_ = ax1.hist(np.random.rand(100), bins = 20, color = 'k', alpha = 0.3)
_
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.plot(np.random.rand(100).cumsum(), 'ko--')
# or 
plt.plot(np.random.rand(1000).cumsum(), linestyle = 'dashed', marker = 'o')
plt.show()
```
