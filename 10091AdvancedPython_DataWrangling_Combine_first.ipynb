```python
# merge
import pandas as pd
df1 = pd.DataFrame({'lkey': ['b','b','c','a','a','d','b'],
                   'data1': range(7)})
df2 = pd.DataFrame({'rkey': ['b','c','d'],
                    'data2': range(3)})
# keys are different in each dataframe by default inner
df3 = pd.merge(df1,df2, left_on = 'lkey', right_on='rkey', how = 'outer')
df3
```

```python
# Combine first
import pandas as pd
import numpy as np
df1 = pd.DataFrame({'a':[np.nan,1,2],
                    'b': [2,3,np.nan],
                    'c': [4,np.nan,5]
                    })
df1

df2 = pd.DataFrame({'a':[np.nan, 2, 4, 5],
                    'b': [2,3,np.nan, 6]})
df2
# combine first
df3 = df1.combine_first(df2)
df3
```
