```python
#Reordering the index
import pandas as pd
import numpy as np
frame = pd.DataFrame(np.arange(12).reshape(4,3), 
                     columns = [['Ohio', 'Colarado', 'Colarado'], ['Green','Red', 'Blue']],
                     index = [['a', 'a', 'b', 'b'], [1,2,1,2]])
frame.index.names = ['key1', 'key2']
frame.columns.names = ['State', 'Color']
frame
# Reordering the indexes 
frame.swaplevel('key1', 'key2')
frame.sort_index(level = 1)
frame.swaplevel(0,1).sort_index(level = 0)
frame.sum(level = 'key2')
```

```python
# merge
import pandas as pd
df1 = pd.DataFrame({'lkey': ['b','b','c','a','a','d','b'],
                   'data1': range(7)})
df2 = pd.DataFrame({'rkey': ['b','c','d'],
                    'data2': range(3)})
# keys are different in each dataframe by default inner
df3 = pd.merge(df1,df2, left_on = 'lkey', right_on='rkey', how = 'outer')
df3
```
