```python
# Leetcode Arrays - 01 Maximum consecutives of 1's
from typing import List
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        count = 0
        result = 0
        for i in range(len(nums)):
            if nums[i] == 1:
                count = count + 1
                result = max(result, count)
            else:
                count = 0
        return result
nums = [1,0,1,1,1,0,1]

obj = Solution()
obj.findMaxConsecutiveOnes(nums)
```
